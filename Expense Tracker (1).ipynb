```python
total=0
listt=[]
while True:
    while True:
        try:
            exp=int(input('enter expense: '))
            break
        except ValueError:
            print("Please enter a valid number.")
    listt.append(exp)
    total+=exp
    while True:
        x=input('Do you want to add another expense?, Enter y else enter n')
        if x.lower()=='y':
            break
        if x.lower()=='n':
            break
        else:
            print('enter a valid number')
            continue
    if x.lower()=='y':
        continue
    if x.lower()=='n':
        break
for i in range(len(listt)):
    print(i+1,'. ',listt[i])
print('TOTAL EXPENSE :', total)
print('AVERAGE EXPENSE :', total/len(listt))
print('MAXIMUM EXPENSE AMOUNT :', max(listt))
print('MINIMUM EXPENSE AMOUNT :', min(listt))
```
